<b>Exercise 2:</b> Write a program to look for lines of the form:

New Revision: 39772

Extract the number from each of the lines using a regular expression and the findall() method. Compute the average of the numbers and print out the average as an integer.

Enter file:mbox.txt<br>
38549

Enter file:mbox-short.txt<br>

```python
import re

fhand = input('Enter file: ')
try:
    fhand = open(fhand)
except:
    print('Error to open a file', fhand)
else:
    lst = list()
    for line in fhand:
        line = line.rstrip()
        x = re.findall('^New Revision:.(\d+)', line)
        if len(x) > 0:
            lst += x
    lst = [int(i) for i in lst]
    print(int(sum(lst)/len(lst)))
```

```python
import re
try:
    fhand = open('regex_sum_373489.txt')
except:
    print('Error to open a file', fhand)
else:
    lst = list()
    for line in fhand:
        line = line.rstrip()
        x = re.findall('\d+', line)
        if len(x) > 0:
            lst += x
    lst = [int(i) for i in lst]
    print(sum(lst))
```

```python
from urllib.request import urlopen
import urllib.parse as par
import json
import ssl

api_key = False
# If you have a Google Places key, enter it here
# api_key = 'AIzaSy___IDByT70'
# https://developers.google.com/maps/documentation/geocoding/intro

if api_key is False:
    api_key = 42
    serviceurl = 'http://py4e-data.dr-chuck.net/json?'
else:
    serviceurl = 'https://maps.googleapis.com/maps/api/geocode/json?'

# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

while True:
    address = input('Enter location: ')
    if not address:  break
    
    parms = dict()
    parms['address'] = address
    if api_key is not False: parms['key'] = api_key
    url = serviceurl + par.urlencode(parms)
    
    print('Retrieving', url)
    uh = urlopen(url, context=ctx)
    data = uh.read().decode()
    print('Retrieved', len(data), 'characters')
    
    try:
        js = json.loads(data)
    except:
        js = None
    
    if not js or 'status' not in js or js['status'] != 'OK':
        print('==== Failure To Retrieve ====')
        print(data)
        continue
    
    print(json.dumps(js, indent=4))
    
    lat = js['results'][0]['geometry']['location']['lat']
    lng = js['results'][0]['geometry']['location']['lng']
    print('lat', lat, 'lng', lng)
    location = js['results'][0]['formatted_address']
    print(location)
    try:
        country_code = js['results'][0]['address_components'][3]['short_name']
    except:
        print('Not have country code')
    else:
        print(country_code)
```

```python
%%time
from urllib.request import urlopen, Request
import urllib.parse, urllib.error
import json
import ssl

# ejemplo = 'http://py4e-data.dr-chuck.net/comments_42.json'
# assigment = 'http://py4e-data.dr-chuck.net/comments_373494.json'

# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input('Enter location:')
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
print('Retrieving ',url)
with urlopen(req, context=ctx) as response:
    html = response.read().decode()
    print('Retrieved', len(html), 'characters')
    try:
        comments = json.loads(html)
    except:
        comments = None
    if not comments:
        print('==== Failure To Retrieve ====')
        print(html)
    suma = 0
    for comment in comments['comments']:
        suma += comment['count']
    print('Count: ',len(comments['comments']))
print('Sum: ',suma)
```
